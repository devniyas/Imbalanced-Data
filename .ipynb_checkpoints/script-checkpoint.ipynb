# Imbalanced Data

The data used in this project is imbalanced. In real life, some extremely critical situations result in imbalanced data sets.ML algorithms tend to tremble when faced with imbalanced classification data sets. Moreover, they result in biased predictions and misleading accuracies. This project trying to address a classification a problem on imbalanced data.

### Data Cleaning

```python
#check missing values in numerical data
num_train.isnull().sum()
```

```python
num_test.isnull().sum()
```

While working on numeric variables, a good practice is to check for correlation in numeric variables. We can remove numeric columns with correlation more than 0.7

```python
num_train_corr = num_train.corr().abs()
num_train_corr = num_train_corr.unstack()
num_train_corr = num_train_corr.order(kind="quicksort")
num_train_corr[(num_train_corr>.7) & (num_train_corr <1)]
```

Removing the column 'weeks_worked_in_year' from both training and testing set.

```python
num_train = num_train[filter(lambda a: a != 'weeks_worked_in_year', num_train.columns)]

# Select columns with more than 45% missing values
temp = cat_train.isnull().mean()
mvtr = temp[temp>0.45].index
temp = cat_test.isnull().mean()
mvte = temp[temp>0.45].index

# Removing the columns with more than 45% missing values
cat_train = cat_train[cat_train.columns.difference(mvtr)]
cat_test = cat_test[cat_train.columns.difference(mvte)]

# Filling NA values with 'Unavailable'
cat_train = cat_train.fillna('Unavailable')
cat_test = cat_test.fillna('Unavailable')
```

#### Data Manipulation
In previous analysis, we saw that categorical variables have several levels with low frequencies. Such levels don’t help as chances are they wouldn’t be available in test set.

```python
def combine_factor(data, threshould):
    """Combining factors with frequency less than threshould value"""
    p = len(data) * threshould / 100
    for i in data.columns:
        temp = data.groupby(i)[i].count()
        factors = temp[temp<p].index.tolist()
        data.loc[data[i].isin(factors), i] = 'Other'
```

```python
# Training data
combine_factor(cat_train, 5)
# Testing data
combine_factor(cat_test, 5)
```

```python
# Calculating the number of factors in each column for training and test
train_factors, test_factors = {}, {}
for col in cat_train:
    train_factors[col] = len(cat_train[col].unique())
    test_factors[col] = len(cat_test[col].unique())
```

```python
# Checking whether the number of factors in each column is same for both training and test data
{k:v for k,v in test_factors.iteritems() if test_factors[k] != train_factors[k] }
```

Before proceeding to the modeling stage, let’s look at numeric variables and reflect on possible ways for binning. Since a histogram wasn’t enough for us to make decision, let’s create simple tables representing counts of unique values in these variables as shown:

```python
num_train['age'].value_counts(sort=False)
```

```python
num_train['wage_per_hour'].value_counts(sort=False)
```

Similarly, you should check other variables also. After this activity, we are clear that more than 70-80% of the observations are 0 in these variables. Let’s bin these variables accordingly. I used a decision tree to determine the range of resultant bins. However, it will be interested to see how 0-25, 26-65, 66-90 works (discerned from plots above).

```python
# bin age variable 0-30 31-60 61 - 90
bins = [0, 30, 60, 90]
labels = ['Young', 'Adult', 'Old']
num_train['age'] = pd.cut(num_train['age'], bins, labels=labels)
num_test['age'] = pd.cut(num_test['age'], bins, labels=labels)
```

```python
from sklearn import preprocessing
encoder = preprocessing.LabelEncoder()
```

```python
# Bin numeric variables with Zero and MoreThanZero
num_train.loc[num_train['wage_per_hour']>0, 'wage_per_hour'] = 'MoreThanZero'
num_train.loc[num_train['wage_per_hour']==0, 'wage_per_hour'] = 'Zero'
num_test.loc[num_test['wage_per_hour']>0, 'wage_per_hour'] = 'MoreThanZero'
num_test.loc[num_test['wage_per_hour']==0, 'wage_per_hour'] = 'Zero'
num_train['wage_per_hour'] = encoder.fit_transform(['wage_per_hour'])
num_test['wage_per_hour'] = encoder.fit_transform(['wage_per_hour'])

num_train.loc[num_train['capital_gains']>0, 'capital_gains'] = 'MoreThanZero'
num_train.loc[num_train['capital_gains']==0, 'capital_gains'] = 'Zero'
num_test.loc[num_test['capital_gains']>0, 'capital_gains'] = 'MoreThanZero'
num_test.loc[num_test['capital_gains']==0, 'capital_gains'] = 'Zero'
num_train['capital_gains'] = encoder.fit_transform(['capital_gains'])
num_test['capital_gains'] = encoder.fit_transform(['capital_gains'])

num_train.loc[num_train['capital_losses']>0, 'capital_losses'] = 'MoreThanZero'
num_train.loc[num_train['capital_losses']==0, 'capital_losses'] = 'Zero'
num_test.loc[num_test['capital_losses']>0, 'capital_losses'] = 'MoreThanZero'
num_test.loc[num_test['capital_losses']==0, 'capital_losses'] = 'Zero'
num_train['capital_losses'] = encoder.fit_transform(['capital_losses'])
num_test['capital_losses'] = encoder.fit_transform(['capital_losses'])

num_train.loc[num_train['dividend_from_Stocks']>0, 'dividend_from_Stocks'] = 'MoreThanZero'
num_train.loc[num_train['dividend_from_Stocks']==0, 'dividend_from_Stocks'] = 'Zero'
num_test.loc[num_test['capital_gains']>0, 'capital_gains'] = 'MoreThanZero'
num_test.loc[num_test['capital_gains']==0, 'capital_gains'] = 'Zero'
num_train['capital_gains'] = encoder.fit_transform(['capital_gains'])
num_test['capital_gains'] = encoder.fit_transform(['capital_gains'])
```
